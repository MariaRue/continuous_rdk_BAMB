```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1-gaussian

def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

x_lin = np.linspace(-5,5,1000)
plt.plot(x_lin, 1-gaussian(x_lin,0,2))
plt.show()
```

```python
#sum of 2 sigmoids

def sigmoid(x, beta, bias):
    return 1/(1+np.exp(-beta*(x+bias)))

x_lin = np.linspace(-5,5,1000)
beta = 2
bias = -4
plt.plot(x_lin, sigmoid(x_lin, beta, bias),label='sig1')
plt.plot(x_lin, sigmoid(-x_lin, beta, bias),label='sig2')
plt.plot(x_lin, sigmoid(x_lin, beta, bias) + sigmoid(-x_lin, beta, bias),label='sum')
plt.legend()
plt.show()
```

It looks like what we need to use is a sum of 2 sigmoids

```python
#abs then sigmoid

def abs_sigmoid(x, beta, bias):
    return 1/(1+np.exp(-beta*(x+bias)))

x_lin = np.linspace(-5,5,1000)
beta = 2
bias = 1
plt.plot(x_lin, sigmoid(np.abs(x_lin), beta, bias))
plt.show()
```

```python
#softmax for 3 classes
def softmax_resp(x, beta, bias):
    return (np.exp(beta*x+bias) + np.exp(-beta*x+bias)) / (1 + np.exp(beta*x+bias) + np.exp(-beta*x+bias))

x_lin = np.linspace(-50,50,1000)
beta = 1
bias = -10
plt.plot(x_lin, softmax_resp(x_lin, beta, bias))
plt.axhline(y=0,color='black',linestyle='--')
plt.show()
```
